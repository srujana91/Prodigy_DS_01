```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'API_SP.POP.TOTL_DS2_en_csv_v2_23.csv'
world_population = pd.read_csv(file_path, skiprows=4)

# Extract relevant data for 2023
population_2023 = world_population[['Country Name', '2023']].dropna()

# Sort by population in descending order and select the top 20 countries
population_2023_sorted = population_2023.sort_values(by='2023', ascending=False).head(20)

# Plot the bar chart
plt.figure(figsize=(12, 8))
plt.barh(population_2023_sorted['Country Name'], population_2023_sorted['2023'], color='skyblue')
plt.xlabel('Population in 2023')
plt.ylabel('Country')
plt.title('Top 20 Countries by Population in 2023')
plt.gca().invert_yaxis()  # Invert y-axis to have the highest population on top
plt.tight_layout()
plt.show()
```
